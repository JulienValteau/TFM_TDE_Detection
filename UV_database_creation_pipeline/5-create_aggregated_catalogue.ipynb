```python
df_entries_galex = df_sources.merge(df_entries_galex,on='SRCNUM_CUV')
```

```python
from pandas.api.types import CategoricalDtype
cat_type = CategoricalDtype(categories=["True","False","<NA>"], ordered=False)
df_entries_galex_NUV = df_entries_galex[['SRCNUM_CUV','SRCNUM_OM','SRCNUM_UVOT','RA', 'DEC','POSERR','COUNT','W1mag', 'W2mag', 'logM',\
                                         'NUV_UVW2_FLUX','NUV_UVW2_FLUX_ERR','NUV_UVM2_FLUX', 'NUV_UVM2_FLUX_ERR','NUV_UVW1_FLUX', \
                                         'NUV_UVW1_FLUX_ERR','SRCNUM_GALEX']].rename(columns={'NUV_UVW2_FLUX':'UVW2_FLUX',\
                                                                                              'NUV_UVW2_FLUX_ERR':'UVW2_FLUX_ERR',\
                                                                          'NUV_UVM2_FLUX':'UVM2_FLUX','NUV_UVM2_FLUX_ERR':'UVM2_FLUX_ERR',\
                                                                          'NUV_UVW1_FLUX':'UVW1_FLUX','NUV_UVW1_FLUX_ERR':'UVW1_FLUX_ERR',\
                                                                          'SRCNUM_GALEX':'OBSID'})
df_entries_galex_FUV=df_entries_galex[['SRCNUM_CUV','SRCNUM_OM','SRCNUM_UVOT','RA', 'DEC','POSERR','COUNT','W1mag', 'W2mag', 'logM',\
                                       'FUV_UVW2_FLUX','FUV_UVW2_FLUX_ERR','FUV_UVM2_FLUX', 'FUV_UVM2_FLUX_ERR','FUV_UVW1_FLUX', \
                                       'FUV_UVW1_FLUX_ERR', 'SRCNUM_GALEX']].rename(columns={'FUV_UVW2_FLUX':'UVW2_FLUX',\
                                                                                             'FUV_UVW2_FLUX_ERR':'UVW2_FLUX_ERR',\
                                                                          'FUV_UVM2_FLUX':'UVM2_FLUX','FUV_UVM2_FLUX_ERR':'UVM2_FLUX_ERR',\
                                                                          'FUV_UVW1_FLUX':'UVW1_FLUX','FUV_UVW1_FLUX_ERR':'UVW1_FLUX_ERR',\
                                                                          'SRCNUM_GALEX':'OBSID'})
for name in ['U_FLUX','U_FLUX_ERR', 'B_FLUX', 'B_FLUX_ERR', 'V_FLUX', 'V_FLUX_ERR']:
    df_entries_galex_NUV[name] = np.nan
    df_entries_galex_FUV[name] = np.nan

for name in ['UVW2_QUALITY_FLAG', 'UVM2_QUALITY_FLAG','UVW1_QUALITY_FLAG','U_QUALITY_FLAG', 'B_QUALITY_FLAG', 'V_QUALITY_FLAG']:
    df_entries_galex_NUV[name] = 0
    df_entries_galex_FUV[name] = 0
    
for name in ['UVW2_EXTENDED', 'UVM2_EXTENDED', 'UVW1_EXTENDED', 'U_EXTENDED','B_EXTENDED', 'V_EXTENDED']:
    df_entries_galex_NUV[name] = "<NA>"
    df_entries_galex_NUV[name] = df_entries_galex_NUV[name].astype(cat_type)
    df_entries_galex_FUV[name] = "<NA>"
    df_entries_galex_FUV[name] = df_entries_galex_FUV[name].astype(cat_type)

df_entries_galex_NUV['MJD'] = 54466
df_entries_galex_FUV['MJD'] = 54466
df_entries_galex_NUV['CAT'] = 'GALEX_NUV'
df_entries_galex_FUV['CAT'] = 'GALEX_FUV'
```

```python
df_entries_galex = pd.concat([df_entries_galex_NUV,df_entries_galex_FUV])
```

```python
df_entries_galex= df_entries_galex[list(df_entries_om.columns.values)]
```
