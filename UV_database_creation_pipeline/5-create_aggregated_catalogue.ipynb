```python
import pandas as pd
import numpy as np
roots_tfm = '/home/julien/Documents/Etudes/Astrofisica/Master/TFM/Data'
```

```python
df_entries_uvot = pd.read_csv(roots_tfm +'/Swift-UVOT/Entries_UVOT_galaxies.csv')
df_entries_om = pd.read_csv(roots_tfm +'/XMM-Newton/Entries_OM_galaxies.csv')
df_entries_galex = pd.read_csv(roots_tfm+'/Galex/Entries_Galex_Galaxies.csv')
df_sources = pd.read_csv(roots_tfm +'/SOURCE_complete_galaxies.csv')
df_sources.head()
```

```python
df_sources[df_sources['COUNT']>2].shape[0]
```

```python
df_entries_om.head()
```

```python
df_sources = df_sources[df_sources['COUNT']>2]
```

```python
df_entries_uvot = df_sources.merge(df_entries_uvot,left_on='SRCNUM_UVOT',right_on='SRCNUM').merge(df_entries_galex,on='SRCNUM_CUV',how='left').drop({'SRCNUM'},axis=1)
df_entries_uvot = df_entries_uvot.loc[:, ~df_entries_uvot.columns.str.contains('^Unnamed')]
df_entries_uvot['CAT'] = 'UVOT'
```

```python
df_entries_uvot.head()
```

```python
df_entries_om = df_sources.merge(df_entries_om,left_on='SRCNUM_OM',right_on='SRCNUM').merge(df_entries_galex,on='SRCNUM_CUV',how='left').drop({'SRCNUM'},axis=1)
df_entries_om = df_entries_om.loc[:, ~df_entries_om.columns.str.contains('^Unnamed')]
df_entries_om['CAT'] = 'OM'
```

```python
df_entries = pd.concat([df_entries_om,df_entries_uvot])
df_entries.sort_values('SRCNUM_CUV').reset_index(drop=True)
df_entries.shape
```

```python
df_entries.to_csv(roots_tfm+'/Entries_galaxies_more_than_2.csv',index=False)
```
